```python

import seaborn as sns
```

```python
import pandas as pd

d = {'X': pd.Series([1, 2, 3,4]), 'Y': pd.Series([1, 2, 3, 4],dtype='int')} 
d = pd.DataFrame(d)
d 
```

```python
import matplotlib.pyplot as plt
sns.lmplot('X','Y',data=d)
plt.show()
```

```python
sns.kdeplot([3,7,9,2])
```

```python
sns.kdeplot([1,3,5,7],[3,3,6,2])
plt.show()
```

```python
sns.factorplot('X','Y',data=d,kind='bar',palette='Greens',size=4,aspect=6)
plt.show()
```

```python
sns.distplot([1,2,3,8,9])
plt.show()
```

```python
sns.distplot([1,2,3,8,9],kde=False,rug=True)
plt.show()
```

```python
sns.rugplot([1,2,3,8,9])
plt.show()
```

```python
sns.countplot([1,2,3,3,3,4,4,5])
plt.show()
```
